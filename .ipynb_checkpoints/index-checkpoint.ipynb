**Whiskers**: The lines extending from the box show the range of data within 1.5 times the IQR. They indicate the spread of most of the data points.

**Outliers (Dots Outside the Whiskers)**: Data points that fall outside 1.5 times the IQR are considered outliers and are shown as individual points. The presence of outliers can indicate unusual data points or potential errors in the dataset.

**Total static head** is a fluid dynamics concept, particularly relevant to pumps and water systems. It represents the total height or pressure a pump must overcome to move fluid from one point to another. This total is the sum of three components:

Elevation Head: The vertical height the fluid must be lifted.

Pressure Head: The pressure difference between the suction and discharge sides of the pump.

Friction Head: The pressure lost due to friction as the fluid moves through the system, including pipes and fittings.

Together, these components determine the pump's power requirements.

# MODEL PERFORMANCE CLASS

```python
# List to store metrics
metrics_list = []

class ModelMetrics:
    def print_model_scores(self, model, y_true, y_pred):
        """
        Print detailed classification metrics for the given model.

        Parameters:
        model: The machine learning model used for predictions.
        y_true: The true labels.
        y_pred: The predicted labels.
        """
        print("Classification Report:\n", classification_report(y_true, y_pred))
        print("Accuracy Score:", accuracy_score(y_true, y_pred))
        print("Precision Score (weighted):", precision_score(y_true, y_pred, average='weighted'))
        print("Recall Score (weighted):", recall_score(y_true, y_pred, average='weighted'))
        print("F1 Score (weighted):", f1_score(y_true, y_pred, average='weighted'))
        
        # Display Confusion Matrix
        cnf_matrix = confusion_matrix(y_true, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cnf_matrix, display_labels=model.classes_)
        print('\nConfusion Matrix:')
        disp.plot()
    
    def collect_metrics(self, model_name, y_true, y_pred):
        """
        Collect and store metrics for a model.

        Parameters:
        model_name: Name of the model.
        y_true: The true labels.
        y_pred: The predicted labels.
        
        Returns:
        A list of dictionaries containing metrics for each model.
        """
        metrics = {
            'model_name': model_name,
            'accuracy': accuracy_score(y_true, y_pred),
            'f1_score': f1_score(y_true, y_pred, average='weighted'),
            'precision': precision_score(y_true, y_pred, average='weighted'),
            'recall': recall_score(y_true, y_pred, average='weighted')
        }
        metrics_list.append(metrics)
        return metrics_list
```

# BASELINE MACHINE LEARNING MODELS

```python
from sklearn.neighbors import KNeighborsClassifier

# Define pipelines
pipe_lr = Pipeline([
    ('scl', StandardScaler(with_mean=False)),
    ('clf', LogisticRegression(random_state=42, max_iter=1000))
])
pipe_svm = Pipeline([
    ('scl', StandardScaler(with_mean=False)),
    ('clf', SVC(random_state=42))
])
pipe_rf = Pipeline([
    ('clf', RandomForestClassifier(random_state=42))
])
pipe_knn = Pipeline([  # Replacing MultinomialNB with KNeighborsClassifier
    ('scl', StandardScaler(with_mean=False)),
    ('clf', KNeighborsClassifier())
])
pipe_dt = Pipeline([
    ('clf', DecisionTreeClassifier(random_state=42))
])

# List of pipelines for easy iteration
pipelines = [pipe_lr, pipe_svm, pipe_rf, pipe_knn, pipe_dt]  # Updated list
pipe_dict = {0: 'Logistic Regression', 1: 'Support Vector Machine',
             2: 'Random Forest', 3: 'K-Nearest Neighbors', 4: 'Decision Tree'}  # Updated dictionary

# Initialize variables to track the best model
best_accuracy = 0
best_pipeline = None
best_model_name = ""

print("Training and evaluating baseline models...\n")
for idx, pipe in enumerate(pipelines):
    # Train the model
    pipe.fit(X_train, y_train)  # No need for special handling anymore

    # Predict on the test set
    y_pred = pipe.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)

    # Calculate precision, recall, and F1 score
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    # Print the classification report
    print(f'{pipe_dict[idx]} pipeline classification report:\n')
    print(classification_report(y_test, y_pred))

    # Print the confusion matrix
    print(f'Confusion Matrix for {pipe_dict[idx]}:\n')
    print(confusion_matrix(y_test, y_pred))

    # Print the precision, recall, F1-score, and accuracy
    print(f'{pipe_dict[idx]} Model Metrics:')
    print(f'  Accuracy: {accuracy:.4f}')
    print(f'  Precision: {precision:.4f}')
    print(f'  Recall: {recall:.4f}')
    print(f'  F1 Score: {f1:.4f}')
    print('-' * 50)

    # Check if this model has the best accuracy
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_pipeline = pipe
        best_model_name = pipe_dict[idx]

# Print the best model details
print(f"\nThe best performing model is {best_model_name} with an accuracy of {best_accuracy:.3f}.")
```

# TUNING THE BEST MODEL

```python
# Define the parameter grid for Random Forest
param_grid_rf = {
    'clf__n_estimators': [50, 75],         # Lower number of trees for quicker runs
    'clf__max_depth': [None, 10],          # Either no limit or a shallow depth to reduce complexity
    'clf__min_samples_split': [5, 10],     # Larger splits reduce the number of nodes in trees
    'clf__min_samples_leaf': [2, 4]        # Larger leaf size reduces complexity
}

# Perform GridSearchCV for Random Forest
print("Performing hyperparameter tuning for Random Forest...\n")
grid_search_rf = GridSearchCV(estimator=best_pipeline, param_grid=param_grid_rf,
                              cv=5, n_jobs=-1, verbose=2, scoring='accuracy')
grid_search_rf.fit(X_train, y_train)

# Get the best parameters and the best model
best_params_rf = grid_search_rf.best_params_
best_tuned_rf_model = grid_search_rf.best_estimator_

# Print the best parameters
print(f"Best parameters for Random Forest: {best_params_rf}\n")

# Evaluate the tuned model on the test set
y_pred_tuned_rf = best_tuned_rf_model.predict(X_test)
accuracy_tuned_rf = accuracy_score(y_test, y_pred_tuned_rf)
# Print the classification report
print("Random Forest tuned model classification report:\n")
print(classification_report(y_test, y_pred_tuned_rf, target_names=['Functional', 'Functional Needs Repair', 'Non Functional']))

# Generate the confusion matrix with custom labels
conf_matrix_rf = confusion_matrix(
    y_test,
    y_pred_tuned_rf,
    labels=["functional", "functional needs repair", "non functional"]
)

# Create a confusion matrix display object with custom labels
disp_rf = ConfusionMatrixDisplay(
    confusion_matrix=conf_matrix_rf,
    display_labels=["Functional", "Functional Needs Repair", "Non Functional"]
)

# Plot the confusion matrix
disp_rf.plot(cmap='Blues')
plt.title("Confusion Matrix for Random Forest Tuned Model")
plt.show()
```

# CONCLUSION 
I observed the following:

**Relationship Between Pump Functionality and Continuous Variables**

1.Amount of Total Static Head (amount_tsh): The distribution of amount_tsh shows overlapping densities across all three target classes. However, the functional pumps exhibit a broader and more consistent distribution, suggesting that a wide range of water head values are associated with functional pumps. Non-functional and functional-needs-repair pumps show more concentrated densities.

2.GPS Height (gps_height): For gps_height, functional pumps are again broadly distributed across various heights, but with noticeable peaks. Non-functional pumps show a narrower distribution, suggesting that non-functional pumps might be more common in certain height ranges. Functional-needs-repair pumps have a more varied distribution.

3.Population: The population variable has a skewed distribution, with most pumps serving smaller populations. Functional pumps tend to serve a wider range of populations, while non-functional pumps are more concentrated in smaller populations. Functional-needs-repair pumps show a similar pattern to functional pumps but with lower densities.

4.Construction Year: The density plot for construction_year indicates that older pumps (constructed earlier) are more likely to be non-functional, while newer pumps tend to be functional. This suggests that the age of the pump might be a factor in its operational status, with older pumps more prone to being non-functional or needing repair.

Overall, these findings suggest that functional pumps are more likely to be found across a broader range of values for these continuous variables, while non-functional and functional-needs-repair pumps tend to be more concentrated within specific ranges. Additionally, the age of the pump (as indicated by construction_year) appears to be a significant factor in its functionality.

 **Relationship Between Pump Functionality and Categorical Variables**

In general, functional pumps are more frequent than non-functional and functional-needs-repair pumps across most categories, likely due to the overall higher number of functional pumps. However, a notable exception is the "never-pay" class in the payment-type category, where non-functional pumps outnumber the others.
